```python
import googlemaps
import re
from pprint import pprint
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
gmaps = googlemaps.Client(key=os.environ['GOOGLEAPI_KEY'])
```

```python
#取得台北市士林區士東路299巷的地理編碼
geocode_result = gmaps.geocode(address = "台南中山北路")
geocode_result
```

```python
#把士東路299巷的經緯度取出來定義為loc，訂一個半徑變數叫rad
loc = geocode_result[0]['geometry']['location']
rad = 100
print(loc)
#用gmaps.places_radar這個代碼來查total有幾個結果
print("台北市士林區士東路299巷"+str(rad)+"公尺的餐廳數量: "+
str(len(gmaps.places_nearby(keyword="restaurant",location=loc, radius=rad)['results'])))
```

```python
# 執行附近搜索
results = gmaps.places_nearby(
    location=loc,
    radius=100,
    keyword='restaurant',
    open_now=False,
    type='restaurant',
    language='zh-TW'
)

# 打印結果
for place in results['results']:
    print(place['name'])
```

```python
# 定義一個格式化電話號碼的函數
def format_phone_number(phone_number):
     # 去掉空格
    phone_number = phone_number.replace(" ", "")
    # 使用正則表達式匹配並格式化電話號碼
    match = re.match(r'^(\d{2})(\d{4})(\d{4})$', phone_number)
    if match:
        return f"{match.group(1)}-{match.group(2)}-{match.group(3)}"
    return phone_number
```

```python
# 抓取每個餐廳的詳細資訊
for place in results['results']:
    place_id = place['place_id']
    # 獲取詳細信息
    place_details = gmaps.place(place_id=place_id, language='zh-TW')
    
    # 打印餐廳的相關信息
    name = place_details['result'].get('name')
    address = place_details['result'].get('formatted_address')
    phone_number = place_details['result'].get('formatted_phone_number')
    rating = place_details['result'].get('rating')
    user_ratings_total = place_details['result'].get('user_ratings_total')
    price_level = place_details['result'].get('price_level')
    website = place_details['result'].get('website')

    # 格式化電話號碼
    if phone_number:
        phone_number = format_phone_number(phone_number)
    
    
    
    print(f"名稱: {name}")
    print(f"地址: {address}")
    print(f"電話號碼: {phone_number}")
    print(f"評分: {rating} ({user_ratings_total} 則評論)")
    print(f"價位等級：{price_level}")
    print(f"網站: {website}")
    print("-" * 100)
    
```

```python
# 定義一個格式化電話號碼的函數
def format_phone_number(phone_number):
    
    if phone_number is None:
        phone_number = "未提供電話號碼"
        return phone_number
    else:
        # 去掉空格
        phone_number = phone_number.replace(" ", "")
        # 使用正則表達式匹配並格式化電話號碼
        match = re.match(r'^(\d{2})(\d{4})(\d{4})$', phone_number)
        phone_number = f"{match.group(1)}-{match.group(2)}-{match.group(3)}"
        return phone_number
    
# 定義是否為None, 自設定為 “未提供”
def price_web(price_level, website):
    if price_level is None:
        price_level = "未提供價位範圍"
    elif price_level == 1:
        price_level = "300 以內"
    elif price_level == 2:
        price_level = "300 至 900"
    elif price_level == 3:
        price_level = "900 至 1,800"
    else:
        price_level = "超過 1800"

    if website is None:
        website = "此餐廳無網頁"

    return price_level, website


# 創建一個空的列表來存儲餐廳信息
restaurant_list = []
number = 0
# 抓取每個餐廳的詳細資訊
for place in results['results']:
    number += 1
    place_id = place['place_id']
    # 獲取詳細信息
    place_details = gmaps.place(place_id=place_id, language='zh-TW')
    
    # 獲取餐廳的相關信息
    name = place_details['result'].get('name')
    address = place_details['result'].get('formatted_address')
    phone_number = place_details['result'].get('formatted_phone_number')
    rating = place_details['result'].get('rating')
    user_ratings_total = place_details['result'].get('user_ratings_total')
    price_level = place_details['result'].get('price_level')
    website = place_details['result'].get('website')
    


    # 格式化電話號碼
    phone_number = format_phone_number(phone_number)

    # 檢查是否為None, 自設定為 未提供
    price_level, website = price_web(price_level, website)
    
    # 創建一個字典來存儲餐廳的信息
    restaurant_info = {
        '編號': number,
        '名稱': name,
        '評分': rating,
        '評論數量': user_ratings_total,
        '價位範圍': price_level,
        '地址': address,
        '電話號碼': phone_number,
        '網站': website
    }
    
    # 將字典添加到列表中
    restaurant_list.append(restaurant_info)

# 美觀地打印餐廳列表
pprint(restaurant_list)
```

```python
import requests

def get_location_by_ip():
    response = requests.get('https://ipinfo.io/')
    data = response.json()
    location = data['loc'].split(',')
    latitude = float(location[0])
    longitude = float(location[1])
    return latitude, longitude

latitude, longitude = get_location_by_ip()

# 創建一個字典來存儲餐廳的信息
locip = {
    'lat': latitude,
    'lng': longitude,
}
print(locip)
```

```python
# 執行附近搜索
ip_results = gmaps.places_nearby(
    location=locip,
    radius=100,
    keyword='restaurant',
    open_now=True,
    type='restaurant',
    language='zh-TW'
)
```

```python

for place in ip_results['results']:
    place_id = place['place_id']

# 獲取詳細信息
place_details = gmaps.place(place_id=place_id, language='zh-TW')
print(type(place_details))
pprint(place_details)
```
